```python
import requests
from bs4 import BeautifulSoup

# Step 1: Fetch the main page content
main_page_url = "https://pokemondb.net/pokedex/national"  # Replace with the URL of the main Pokemon page
response = requests.get(main_page_url)
main_page_content = response.content

print(response.content)
```

```python
# Step 2: Print the prettified content
soup = BeautifulSoup(main_page_content, "html.parser")
prettified_content = soup.prettify()
print(prettified_content)
```

```python
# Find all <a> elements with class "ent-name" and extract the text (i.e., the Pokémon names)
pokemon_name_elements = soup.find_all("a", class_="ent-name")
pokemon_names = []

for element in pokemon_name_elements:
    pokemon_names.append(element.text)

# Step 3: Print the extracted Pokémon names
print(len(pokemon_names))
```

```python
pokemon_links=[]
for element in pokemon_name_elements:
    name = element.text
    url = "https://pokemondb.net" + element["href"]
    pokemon_links.append((name, url))
```

```python
print(pokemon_links)
```

```python
response = requests.get(pokemon_links[11][1])
pokemon_page_content = response.content
# Step 2: Print the prettified content
soup = BeautifulSoup(pokemon_page_content, "html.parser")
prettified_content = soup.prettify()
print(prettified_content)
```

```python
import time
# Step 3: Loop through each Pokémon's page and extract the description
pokedex_entries=[]
for name, url in pokemon_links:
    print(f"Fetching information for {name} from {url}")
    response = requests.get(url)
    pokemon_page_content = response.content

    # Parse the individual Pokémon's page content
    pokemon_soup = BeautifulSoup(pokemon_page_content, "html.parser")

    # Find the second <p> element (index 1) containing the description and extract the text
    description_elements = pokemon_soup.find_all("p")
    description = description_elements[1].get_text(strip=True) if len(description_elements) >= 2 else "Description not found."

    # Print the extracted description
    # print("Description:", description)
    pokedex_entries.append((name, description))
    time.sleep(0.05)
```
